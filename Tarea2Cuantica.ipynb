```python
import numpy as np
import math as m
import matplotlib.pyplot as plt

def Cn(n):
    return -m.sqrt(60)*(m.pi*n*m.sin(m.pi*n)+2*m.cos(m.pi*n)-2)/(m.pi**3*n**3)

datos = []

A = m.sqrt(30)
a = 1

for x in np.arange(0, a, 0.01):
    suma = 0
    for n in range(1, 1000):
        suma += Cn(n)*m.sin(n*m.pi*x/a)*m.sqrt(2)
    datos.append(suma)

datos = np.array(datos)
rango = np.arange(0, a, 0.01)

print(datos)

plt.plot(rango, datos)
plt.show()


```
