Para empezar, se graficará la función para t=0 y considerando un pozo potencial con a = 1. Teniendo en cuenta que la parte temporal de la función de onda es de la forma $$ e^{-it/\hbar} $$ , esta será igual a 1 para t=0. Por lo tanto, la función de onda para t=0 será igual a la parte espacial de la función de onda, que es $$\sum C_n \sqrt{\frac{2}{a}}\sin(\frac{n \pi x}{a}) $$

```python
import numpy as np
import math as m
import matplotlib.pyplot as plt

def Cn(n):
    return -m.sqrt(60)*(m.pi*n*m.sin(m.pi*n)+2*m.cos(m.pi*n)-2)/(m.pi**3*n**3)

datos = []


A = m.sqrt(30)
a = 1

print(len(np.arange(0, a, 0.01)))

for x in np.arange(0, a, 0.01):
    suma = 0
    for n in range(1, 1000):
        suma += Cn(n)*m.sin(n*m.pi*x/a)*m.sqrt(2)
    datos.append(suma)





datos = np.array(datos)
rango = np.arange(0, a, 0.01)

plt.plot(rango, datos)
plt.plot(rango, abs(datos)**2)
plt.show()

```

Como podemos observar, el resultado es el esperado. Pues, recordemos que se sabe que para t=0, la función de onda propuesta es
$$ \Psi(x, t=0) = Ax(1-x) $$ 
donde A, como mostramos en el punto 4 parte a, es igual a 
$$ A = \sqrt{30} $$ 
para un a = 1.
