Para empezar, se graficar치 la funci칩n propuesta en la primera perte del problema para t=0, utilizando el $ C_n $ y considerando un pozo potencial con a = 1. Es decir, $$ Ax(1-x) = \sum C_n \sqrt{\frac{2}{a}}\sin(\frac{n \pi x}{a})  $$ 

```python
import numpy as np
import math as m
import matplotlib.pyplot as plt

def Cn(n):
    return -m.sqrt(60)*(m.pi*n*m.sin(m.pi*n)+2*m.cos(m.pi*n)-2)/(m.pi**3*n**3)

datos = []


A = m.sqrt(30)
a = 1

print(len(np.arange(0, a, 0.01)))

for x in np.arange(0, a, 0.01):
    suma = 0
    for n in range(1, 1000):
        suma += Cn(n)*m.sin(n*m.pi*x/a)*m.sqrt(2)
    datos.append(suma)





datos = np.array(datos)
rango = np.arange(0, a, 0.01)

plt.plot(rango, datos)
plt.show()

```

Como podemos observar, el resultado es el esperado. Pues, recordemos que se sabe que para t=0, la funci칩n de onda propuesta es
$$ \Psi(x, t=0) = Ax(1-x) $$ 
donde A, como mostramos en el punto 4 parte a, es igual a 
$$ A = \sqrt{30} $$ 
para un a = 1.





Ahora bien, empezaremos a hacer gr치ficas para diferentes n y considerando el tiempo. 

Para esto, tenemos que 

$$ \Psi_n(x,t) = C_n  \sqrt{\frac{2}{a}}\sin(\frac{n \pi x}{a})e^{-\frac{i  E_n t}{\hbar}}$$

Haremos una gr치fica para la funci칩n de onda con t = 1 y valores de n iguales a 2, 3, 4, 5. La gr치fica de n=1 no se har치, pues su valor es sustancialmente mayor a los dem치s, por lo que no se apreciar칤an bien la gr치fica.

```python
import numpy as np
import math as m
import matplotlib.pyplot as plt

def Cn(n):
    return -m.sqrt(60)*(m.pi*n*m.sin(m.pi*n)+2*m.cos(m.pi*n)-2)/(m.pi**3*n**3)

datosn1 = []
datosn2 = []
datosn3 = []
datosn4 = []
datosn5 = []

#Usamos uniades naturales y asumimos una masa muy peque침a, la de un electron 9.1093837 칑 10-31 kg


def parte_temporal(n, t):
    E_1 = 1/(8*9.12E-31)
    E_n = (n**2)*E_1

    return np.cos(E_n*t) - 1j*np.sin(E_n*t)
    



    

A = m.sqrt(30)
a = 1
n = 1
t = 1

for x in np.arange(0, a, 0.01):
    x_n = m.sin(n*m.pi*x/a)*m.sqrt(2)*parte_temporal(n, t)*Cn(n)
    datosn1.append(x_n)

n = 2 
t = 1

for x in np.arange(0, a, 0.01):
    x_n = m.sin(n*m.pi*x/a)*m.sqrt(2)*parte_temporal(n, t)*Cn(n)
    datosn2.append(x_n)

n = 3 
t = 1

for x in np.arange(0, a, 0.01):
    x_n = m.sin(n*m.pi*x/a)*m.sqrt(2)*parte_temporal(n, t)*Cn(n)
    datosn3.append(x_n)

n = 4 
t = 1

for x in np.arange(0, a, 0.01):
    x_n = m.sin(n*m.pi*x/a)*m.sqrt(2)*parte_temporal(n, t)*Cn(n)
    datosn4.append(x_n)

n = 5 
t = 1

for x in np.arange(0, a, 0.01):
    x_n = m.sin(n*m.pi*x/a)*m.sqrt(2)*parte_temporal(n, t)*Cn(n)
    datosn5.append(x_n)




datosn1 = np.array(datosn1)
datosn2 = np.array(datosn2)
datosn3 = np.array(datosn3)
datosn4 = np.array(datosn4)
datosn5 = np.array(datosn5)
rango = np.arange(0, a, 0.01)




plt.plot(rango, datosn2, label = "n = 2")
plt.plot(rango, datosn3, label = "n = 3")
plt.plot(rango, datosn4, label = "n = 4")
plt.plot(rango, datosn5, label = "n = 5")
plt.legend()
plt.title("Funci칩n de onda para distintos n con t = 1")
plt.show()


plt.plot(rango, abs(datosn2), label = "p(n) con n = 2")
plt.plot(rango, abs(datosn3), label = "p(n) con n = 3")
plt.plot(rango, abs(datosn4), label = "p(n) con n = 4")
plt.plot(rango, abs(datosn5), label = "p(n) con n = 5")
plt.legend()
plt.title("Distribuci칩n de probabilidad para distintos n con t = 1")
plt.show()

```

Notemos que para t=1, la funci칩n de onda con n=2 y n= 4 es una l칤nea en el eje horizontal. Tiene sentido, pues nuestro $C_n $ da 0 cuando los n son pares.
